Exercise 2: The Titanic dataset


```python
# 2(a) Load the titanic.csv dataset into a pandas dataframe named titanic_df.

titanic_df = pd.read_csv('titanic.csv')
titanic_df.head()
```

```python
# 2(b) The column names sibsp and parch are not very descriptive. Use df.rename() to rename these columns to siblings_spouses 
# and parents_children respectively.

titanic_df.rename(columns={'sibsp' : 'siblings_spouses', 'parch' : 'parents_children'}, inplace=True)
titanic_df.head()
```

```python
# 2(c) We will practice indexing different subsets of the dataframe in the following questions. 
# Select the column age using single bracket notation []. What type of object is returned?

titanic_df['age'].info()
```

```python
# 2(d) Now select the age using double bracket notation [[]]. What type of object is returned?

titanic_df[['age']].info()
```

```python
# 2(e) Select the columns pclass, survived, and age using a single line of code.
titanic_df[['pclass', 'survived', 'age']]
```

```python
# 2(f) Use the iloc method to obtain the first 5 rows of the columns name, sex and age using a single line of code.
titanic_df.iloc[0:5, 2:5]
```

```python
# 2(g) Now use the loc method to obtain the first 5 rows of the columns name, sex and age using a single line of code.
titanic_df[['name', 'sex', 'age']].loc[0:5]
```

```python
# 2(h) How many passengers survived (survived = 1) the disaster? Hint: try using df.query() or [] notation to subset the dataframe 
# and then df.shape to check its size.
survived_passengers = titanic_df.loc[titanic_df.survived == 1]
print(survived_passengers)
survived_passengers.shape
```

```python
# 2(i) How many passengers that survived the disaster (survived 1) were over 60 years of age?
titanic_df.loc[(titanic_df.survived == 1 ) & (titanic_df.age >= 60)]
```

```python
# 2(j) What was the lowest and highest fare paid to board the titanic? Store your answers as floats in the variables lowest and highest.
highest_fair_paid = np.max(titanic_df.fare)
lowest_fare_paid = np.min(titanic_df.fare)
print(f'The highest fare paid is {highest_fair_paid} \nThe lowest fare paid is {lowest_fare_paid}')
```

```python
# 2(k) Sort the dataframe by fare paid (most to least).
titanic_df.sort_values(by='fare')
```
